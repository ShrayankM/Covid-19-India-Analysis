#Main Covid-19 Data

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("covid_19_india.csv")
```

```
data.head()
```

#Cases Plots

```
states = data['State/UnionTerritory']
cases = data['Confirmed']
```

```
states_set = set(states)
states_set = sorted(states_set)
states_set.remove('Unassigned')
print(states_set)
```

```
states_cases = dict()
```

```
for item in states_set:
    case_count = np.array(cases[states == item])
    states_cases[item] = case_count[-1]
```

```
print(states_cases)
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
states = states_cases.keys()
cases = states_cases.values()
ax.bar(states,cases,color='orange')
plt.xlabel('States', fontsize=20)
plt.ylabel('No. of Cases', fontsize=20)
plt.xticks(rotation=90)
plt.gcf().set_size_inches([10,8])
plt.show()
```

```
states_case_copy = states_cases.copy()
sorted_cases = sorted(states_case_copy.items(),key = lambda item:item[1],reverse=True)
```

```
print(sorted_cases)
```

```
states_list = list()
case_list   = list()
for item in sorted_cases:
    states_list.append(item[0])
    case_list.append(item[1])
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
states = states_list
cases = case_list
ax.bar(states,cases,color='orange')
plt.xlabel('States', fontsize=20)
plt.ylabel('No. of Cases', fontsize=20)
plt.xticks(rotation=90)
plt.gcf().set_size_inches([10,8])
plt.show()
```

#Cured Plots

```
states = data['State/UnionTerritory']
cured = data['Cured']
```

```
states_set = set(states)
states_set = sorted(states_set)
states_set.remove('Unassigned')
```

```
states_cured = dict()
```

```
for item in states_set:
    cured_count = np.array(cured[states == item])
    states_cured[item] = cured_count[-1]
```

```
print(states_cured)
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
states = states_cured.keys()
cured = states_cured.values()
ax.bar(states,cured,color='green')
plt.xlabel('States', fontsize=20)
plt.ylabel('No. of Cured', fontsize=20)
plt.xticks(rotation=90)
plt.gcf().set_size_inches([10,8])
plt.show()
```

```
states_cured_copy = states_cured.copy()
sorted_cured = sorted(states_cured_copy.items(),key = lambda item:item[1],reverse=True)
```

```
print(sorted_cured)
```

```
states_list = list()
cured_list   = list()
for item in sorted_cured:
    states_list.append(item[0])
    cured_list.append(item[1])
```

```
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
states = states_list
cured = cured_list
ax.bar(states,cured,color='green')
plt.xlabel('States', fontsize=20)
plt.ylabel('No. of Cured', fontsize=20)
plt.xticks(rotation=90)
plt.gcf().set_size_inches([10,8])
plt.show()
```

#Total Cases/Cured

```

N = len(states_set)
cases = states_cases.values()

ind = np.arange(N)  
width = 0.35       

fig = plt.figure()
ax = fig.add_subplot(111)
rects1 = ax.bar(ind, cases, width, color='royalblue')

cured = states_cured.values()

rects2 = ax.bar(ind+width, cured, width, color='seagreen')
ax.set_xlabel('States')
ax.set_ylabel('Cases/Cured')
ax.set_title('Cured and Cases')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels( states_cured.keys(),rotation = 90)

ax.legend( (rects1[0], rects2[0]), ('Cases', 'Cured') )
plt.gcf().set_size_inches([10,8])
plt.show()
```
